```python
import numpy as np
from numpy import linalg as LA
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#questoon 2 a
np.set_printoptions(formatter={'float_kind':lambda x: "%.4f" % x})
arr = np.array([[10,60,10,90],[20,50,40,70],[30,50,30,40],[20,50,20,60],[10,60,30,10]])
normVector = []
for i in range(0,5):
    normVector.append(LA.norm(arr[i]))
    print(normVector[i])
    
```

```python
#question 2 b 
cosine = np.empty((5, 5))
for i in range(0,5):
    for j in range(0,5):
        dot = np.dot(arr[i], arr[j])
        norma = np.linalg.norm(arr[i])
        normb = np.linalg.norm(arr[j])
        cosine[i][j] = dot / (norma * normb)
print(cosine)
```

```python
#question 2 c
distance = np.empty((5, 5))
for i in range(0,5):
    for j in range(0,5):
        distance[i][j] = np.linalg.norm(arr[i]-arr[j])
print(distance)
```

```python
#question 3 c 
x = [10,20,30,20,10]
y = [60,50,50,50,60]
z = [90,70,40,60,10]
meanarray = np.empty(3)
meanarray[0] = np.mean(x)
meanarray[1]= np.mean(y)
meanarray[2]= np.mean(z)
print(meanarray)
```

```python
#question 3  d 
for i in range(0,5):
    for j in range(0,3):
        arrnew[i][j]=arrnew[i][j]-meanarray[j]
print(arrnew)        
```

```python
#question 3 e
arrtranspose=np.transpose(arrnew)
covariance_matrix = np.cov(arrtranspose)
print(covariance_matrix)
```

```python
#question 3 f
eigenvalues, eigenvectors = np.linalg.eigh(covariance_matrix)
eigenvalues= np.diag(eigenvalues)
print(eigenvalues)
print(eigenvectors)
```

```python
#question 3 g
output =  np.matmul(np.matmul(eigenvectors,np.transpose(eigenvalues)),eigenvectors)
print(output)
```
